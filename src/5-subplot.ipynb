### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # 2 rows, 2 columns

### 1st Subplot: Line Chart ###
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X-axis")
axes[0, 0].set_ylabel("Y-axis")

### 2nd Subplot: Bar Chart ###
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axes[0, 1].set_title("Bar Chart")

### 3rd Subplot: Scatter Plot ###
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='purple')
axes[1, 0].set_title("Scatter Plot")

### 4th Subplot: Pie Chart ###
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
